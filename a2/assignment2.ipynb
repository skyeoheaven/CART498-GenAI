```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
import string

# load model + tokenizer
tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")
model.eval()

prompt = "Nothing that is not there and the nothing that"

# tokenize input
inputs = tokenizer(prompt, return_tensors="pt")

# get model output
with torch.no_grad():
    outputs = model(**inputs)

# get logits for the next token only
next_token_logits = outputs.logits[0, -1]

# convert to probabilities
probs = torch.softmax(next_token_logits, dim=0)

# sort tokens by probability (descending)
sorted_probs, sorted_ids = torch.sort(probs, descending=True)

valid_words = []

for token_id in sorted_ids:
    token = tokenizer.convert_ids_to_tokens(token_id.item())

    # remove GPT-2 space marker
    token = token.replace("Ġ", "")

    # keep only alphabetic "word-like" tokens
    if token.isalpha():
        valid_words.append(token)

    if len(valid_words) == 300:
        break

print("nth most probable next word:", valid_words[299])
```
